```python
# Function that calls ChatGPT
def get_completion(prompt, model="gpt-3.5-turbo"):
  messages = [{"role": "user", "content": prompt}]
  response = openai.ChatCompletion.create(model=model, messages=messages, temperature=0,)
  return response.choices[0].message["content"]
```

```python
# Sample prompt and response
prompt = "<What is life like in Norway?>"
response = get_completion(prompt)
print(response)
```

```python
# Uploads list of countries
countries_list = pd.read_csv("/content/Countries-Continents.csv")
```

```python
countries_list
```

```python
# Generates ChatGPT response on what is it like to live in every country
response_list = []
for i in range(len(countries_list)):
  country = countries_list.iloc[i,1]
  prompt = "<What is life like in {}?>".format(country)
  response = get_completion(prompt)
  response_list.append(response)
```

```python
response_list
```

```python
# Adds ChatGPT responses to dataframe
countries_list['Response'] = response_list
```

```python
from textblob import TextBlob
```

```python
# Performs sentiment analysis for each country
review_polarity = []
review_subjectivity = []
for i in range(len(countries_list)):
  review = TextBlob(countries_list.iloc[i,2])
  review_polarity.append(review.sentiment[0])
  review_subjectivity.append(review.sentiment[1])
```

```python
countries_list['Polarity'] = review_polarity # Polarity is how positive the sentiment is from 0 to 1
countries_list['Subjectivity'] = review_subjectivity # Subjectivity is how biased the response is from 0 to 1
```
